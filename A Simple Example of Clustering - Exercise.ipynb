## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
#Load the data
raw_data = pd.read_csv(r'C:\Users\Dell\Documents\Countries-exercise.csv')
data = raw_data.copy()
data
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
x = data.iloc[:,1:3]
x
```

## Clustering

Assume there are only two clusters. 

```python
# with 7 cluster the value of K is 7
kmeans = KMeans(7)
```

```python
kmeans.fit(x)
```

### Clustering Resutls

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
# Add the cluster in the given data
data_with_clusters = data.copy()
data_with_clusters ['Cluster'] = identified_clusters
data_with_clusters
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
# As clustering is the irretative process so we define 'c' parameter with the cluster variable
plt.scatter(data['Longitude'], data['Latitude'],c=data_with_clusters['Cluster'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```
