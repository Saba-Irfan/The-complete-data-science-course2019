# A Simple Example of Clustering 

## Import the relevant libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

## Load the data

Load data from the csv file: <i> 'Countries.csv'</i>.


```python
#Load the data
raw_data = pd.read_csv(r'C:\Users\Dell\Documents\Countries-exercise.csv')
data = raw_data.copy()
data
```

## Plot the data

Plot the <i>'Longtitude'</i> and <i>'Latitude'</i> columns. 

```python
plt.scatter(data['Longitude'],data['Latitude'])
plt.xlim(-180,180)
plt.ylim(-90,90)
plt.show()
```

## Select the features

Create a copy of that data and remove all parameters apart from <i>Longitude</i> and <i>Latitude</i>.

```python
x = data.iloc[:,1:3]
x
```

## Clustering

Assume there are only two clusters. 

```python
# with 7 cluster the value of K is 7
kmeans = KMeans(7)
```

```python
kmeans.fit(x)
```

### Clustering Resutls

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
# Add the cluster in the given data
data_with_clusters = data.copy()
data_with_clusters ['Cluster'] = identified_clusters
data_with_clusters
```

Did you remember to use the <i> c </i> parameter to separate the data by the clusters we defined?

```python
# As clustering is the iterative process so we define 'c' parameter with the cluster variable
plt.scatter(data['Longitude'], data['Latitude'],c=data_with_clusters['Cluster'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

# Selecting the optimal number of cluster


# # WCSS

```python
# use the sklearn method "interia_"
kmeans.inertia_
```

```python
# to apply Elbow Method for finding the optimal value of number of cluster 'K', use  a loop
wcss = []
# 'cl_num' is a that keeps track the highest number of clusters we want to use the WCSS method for.
# Note that 'range' doesn't include the upper boundery
cl_num = 11
for i in range (1,cl_num):
    kmeans= KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

# The Elbow Method

```python
number_clusters = range(1,cl_num)
plt.plot(number_clusters, wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

From the above plot we find out that the optimal number of clusters for this data would be 2 or 3.

```python
# Repeat the same code for clustering with 'K = 2'
kmeans = KMeans(2)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
```

```python
plt.scatter(data['Longitude'], data['Latitude'], c=data_with_clusters['Cluster'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

```python
# Again repeat the code with 'K = 3'
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
```

```python
plt.scatter(data['Longitude'], data['Latitude'], c=data_with_clusters['Cluster'], cmap = 'rainbow')
plt.xlim(-180,180)
plt.ylim(-90, 90)
plt.show()
```

From the above plot of clustering with 'K = 2' and 'K = 3' , we can clearly see that we have the best clustering solution to our data. However, with 'K = 7' we did not get the clear clustering solution.
