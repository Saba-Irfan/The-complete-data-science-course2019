## Create the regression

### Declare the dependent and the independent variables

```python
y=data ['price']
x1= data ['size']
```

### Explore the data

```python
plt.scatter(x1,y)
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```

The above plots shows the a causal relationship exist between 'price' and 'size' of houses, so we can move forward to creat our regression model

### Plot the regression line on the initial scatter

```python
x=sm.add_constant(x1)
results= sm.OLS(y,x).fit()
results.summary()
```

The simple linear regression line follows the eqaution 
yhat = b0 + b1* x1
from the above OLS regression table we find the 
b0 = 101900 and b1 = 223.1787
moreover, the p-value of variable size is 0 that show that size has siginficant effect on the price of houses

```python
plt.scatter(x1,y)
yhat= 223.1787*x1 + 101900
fig = plt.plot(x1,yhat, lw=2, c='orange', label='Regression line')
plt.xlabel('size', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.show()
```
